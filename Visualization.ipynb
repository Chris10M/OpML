```python
## Training loss plot for gaussianSampled

import matplotlib.pyplot as plt
import csv
import numpy as np
  
a_x = []
aw_x = []
aqh_x = []
ap_x = []
as_x = []
an_x = []
afac_x = []
mg_x = []
ng_x = []
ahd_x = []
aopahd_x = []

with open('results/discrete/adam/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      a_x.append(float(col[1]))

with open('results/discrete/Adam_w/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      aw_x.append(float(col[1]))

with open('results/discrete/QHAdam/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      aqh_x.append(float(col[1]))

with open('results/discrete/Adam_P/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      ap_x.append(float(col[1]))

with open('results/discrete/AdamS/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      as_x.append(float(col[1]))

with open('results/discrete/Nadam/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      an_x.append(float(col[1]))

with open('results/discrete/Adafactor/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      afac_x.append(float(col[1]))

with open('results/discrete/MADGRAD/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      mg_x.append(float(col[1]))

with open('results/discrete/NvNovoGrad/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      ng_x.append(float(col[1]))

with open('results/discrete/adam_Hd/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      ahd_x.append(float(col[1]))

with open('results/discrete/adam_op_adam_hd/train_loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for col in plots:
      aopahd_x.append(float(col[1]))

y = [i for i in range(1,len(a_x)+1)]
#plt.yscale('log' )
plt.plot(y,a_x,label = 'Adam')
plt.plot(y,aw_x,label = 'Adam_W')
plt.plot(y,aqh_x,label = 'Adam_QH')
plt.plot(y,ap_x,label = 'Adam_P')
plt.plot(y,as_x,label = 'Adam_S')
plt.plot(y,an_x,label = 'Adam_N')
plt.plot(y,afac_x,label = 'Adam_AdaFactor')
plt.plot(y,mg_x,label = 'madgrad')
plt.plot(y,ng_x,label = 'nvnovograd')
plt.plot(y,ahd_x,label = 'Adam_HD')
plt.plot(y,aopahd_x,label = 'Adam_op_Adam_HD')



plt.xticks(np.arange(1, len(a_x)+1, 1.0))

plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Train loss')
plt.legend()

plt.savefig('discrete_Vae_train_loss.png')
plt.show()
```
